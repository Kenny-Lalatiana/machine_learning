## Implementation Logistic Regression

```python
def sigmoid(x):
    sigma = 1/(1+np.exp(-x))
    return sigma
```

```python
def log_reg(X, y, learning_rate, iterations):
    m = X.shape[0]
    n = X.shape[1]

    W = np.zeros((n, 1))
    b = 0

    cost_list = []

    for i in range(iterations):
        Z = np.dot(X,W) + b 
        A = sigmoid(Z)
        A = np.clip(A, 1e-10, 1 - 1e-10)

        # Calcul de la fonction de coût
        cost_function = -(1/m) * np.sum((y * np.log(A) + (1 - y) * np.log(1 - A)))

        # Calcul des gradients
        dW = (1/m) * np.dot(X.T, (A - y))
        db = (1/m) * np.sum(A - y)

        # Mise à jour des paramètres
        W = W - learning_rate*dW
        b = b - learning_rate*db

        cost_list.append(cost_function)
        if (i%(iterations/10)==0):
            print ("cost after",i,"iterations is",cost_function)

    plt.plot(np.arange(iterations),cost_list)
    plt.xlabel('iterations')
    plt.ylabel('cost value')
    plt.title('COST FUNCTION OVER ITERATIONS')
    plt.show()

    return W, b
```

```python
W , b = log_reg(X_train_split_scaled,y_train_split,0.01,1000)
```

```python
def predict(X, W, b):
    Z = np.dot(X, W) + b
    A = 1 / (1 + np.exp(-Z))
    return (A > 0.5).astype(int)
```

```python
def f1_score(y_true, y_pred):
   
    # Convertir en numpy array 
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    
    # Calcul des vrais positifs, faux positifs et faux négatifs
    TP = np.sum((y_true == 1) & (y_pred == 1))
    FP = np.sum((y_true == 0) & (y_pred == 1))
    FN = np.sum((y_true == 1) & (y_pred == 0))

    # Calcul de la précision et du rappel
    precision = TP / (TP + FP) if (TP + FP) > 0 else 0
    recall = TP / (TP + FN) if (TP + FN) > 0 else 0

    # Calcul du F1-score
    if (precision + recall) > 0:
        f1 = 2 * (precision * recall) / (precision + recall)
    else:
        f1 = 0

    return f1
```

```python
y_pred = predict(X_valid_split_scaled, W, b)

# Calcul de l'accuracy
accuracy = np.mean(y_pred == y_valid_split)
print(f"Accuracy: {accuracy}")

# Calcul du F1-score
f1 = f1_score(y_valid_split, y_pred)
print(f"F1 Score: {f1:.4f}")
```
